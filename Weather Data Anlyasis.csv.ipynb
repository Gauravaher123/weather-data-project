## How to Analyze DataFrames ?

## .head

It shows the first N rows in the data (by default, N=5).

```python
data.head()
```

## .Shape

It Shows the total number of rows and number of columns of the dataframe.

```python
data.shape
```

## .index 

This attribute provides the index of the dataframe.

```python
data.index
```

## .column

It shows the name of each column.

```python
data.columns
```

## .dtypes

It shows the Data-type of each column.

```python
data.dtypes
```

## .unique()

In a column, Itsowes all the values. It can be applied on a single column only. not on the whole dataframe. 

```python
data['Weather'].unique()
```

## .nunique()

It shows the total number of unique values In each column. In can be applied on a single column as well 
as on whole dataframe.

```python
data.nunique()
```

## .count

It shows the total number of non-null values in each column. It can be applied on a single column as well as on 
whole datarame.

```python
data.count()
```

## .value_counts

In a column, it shows the unique values with thier count it can be applied on single column only.

```python
data['Weather'].value_counts()
```

## .info()

Provides basic information about the dataframe

```python
data.info()
```
